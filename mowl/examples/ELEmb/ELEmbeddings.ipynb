# ELEmbeddings

## Imports

```python
import sys
import os
sys.path.append('../../../')
from mowl.datasets.ppi_yeast import PPIYeastDataset
from mowl.embeddings.elembeddings.model import ELEmbeddings

import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np
```

## Loading the dataset

```python
dataset = PPIYeastDataset()
```

## ELEmbeddings

```python
model = ELEmbeddings(dataset)
```

Train the model

```python
model.train()
```

```python
keys, embeddings = model.get_embeddings()
```

```python
X = TSNE(n_components=2, verbose=1, n_iter=2500).fit_transform(embeddings)
```

```python
ec_numbers = {}
with open('../../../data/yeast_ec.tab') as f:
    next(f)
    for line in f:
        it = line.strip().split('\t', -1)
        if len(it) < 5:
            continue
        if it[3]:
            prot_id = it[3].split(';')[0]
            prot_id = '{0}'.format(prot_id)    
            ec_numbers[prot_id] = it[4]
classes = {'0': [[], []]}
for v, i in keys.items():
    if v.startswith('GO_'):
        continue
    v = '4932.' + v
    if v in ec_numbers:
        ec = ec_numbers[v].split('.')[0]
        if ec not in classes:
            classes[ec] = [[], []]
        classes[ec][0].append(X[i, 0])
        classes[ec][1].append(X[i, 1])
        
colors = iter(plt.cm.rainbow(np.linspace(0, 1, len(classes))))
fig, ax = plt.subplots(figsize=(20, 20))

for ec, items in classes.items():
    if ec == '0':
        continue
    color = next(colors)
    ax.scatter(items[0], items[1], color=color, label=ec)

ax.legend()
ax.grid(True)

plt.show()
```
